```python
import pandas as pd
import seaborn as sb
import plotly.express as px #importing all of the necessary packages.
```

```python
#getting our data in

df = pd.read_csv('data/diabetes.csv') #https://www.kaggle.com/shantanudhakadd/diabetes-dataset-for-beginners
df #0 is no diabtes 1 is diabetes
```

```python
dataplot = sb.heatmap(df.corr() , annot=True)
```

```python
#ploting only pataient with diabetes so the outcome should be 1
fig = px.scatter_3d((df[df.Outcome == 1]), x = 'Age' , y = 'BloodPressure', z = 'Glucose', color = 'BMI', opacity=0.6 ,
title = 'Glucose, Blood Pressure, Age and then colored by BMI')
fig.update_traces(marker = dict( size=3.5 ))
fig.show()
```

```python
from sklearn.preprocessing import StandardScaler
features = df[['Age', 'BMI', 'Glucose', 'SkinThickness','Pregnancies', 'BloodPressure', 'Insulin','DiabetesPedigreeFunction' ]]
scaled_features = StandardScaler().fit_transform(features)

from sklearn.decomposition import PCA 
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(scaled_features)
```

```python
fig = px.scatter(scaled_features)
fig
```

```python
fig = px.scatter(principalComponents)
fig
```
